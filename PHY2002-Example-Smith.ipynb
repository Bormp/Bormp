```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,70,500)
y = np.sin(x)
plt.plot(x,y, 's', color = 'black', label = 'yes')
plt.xlabel('hours')
plt.ylabel('miles')
plt.legend(loc = 'best', numpoints = 1)
```

```python
x2 = np.linspace(0,50,500)
y2 = np.cos(x2)

plt.plot(x2, y2, 'd', color= 'green', label = 'miles per hour')
plt.xlabel('hours')
plt.ylabel('miles')
plt.legend(loc = 'best', numpoints = 1)
```
